##Downsample and crop to center on image(right thing to do?)

```python
# subtracting overall mean--remain sensitive to colors
im = imread('../data/0100448_RP-P-OB-80.995B.jpg')
print im.dtype, np.mean(im).dtype
im -= np.mean(im)
im = Image.fromarray(im)
thumb = ImageOps.fit(im, (96,96))
plt.imshow(thumb,vmin=0,vmax=2)
```

```python
#subtracting channel by channel; fewer outliers but not sensitive to color.
im = imread('../data/0100448_RP-P-OB-80.995B.jpg')
im -= np.mean(np.mean(im,axis=0),axis=0)
im = Image.fromarray(im)
thumb = ImageOps.fit(im, (96,96))
plt.imshow(thumb)
```

##Need to subtract mean image before downsampling to avoid artifacts

```python
im = Image.open('../data/0100448_RP-P-OB-80.995B.jpg')
thumb = ImageOps.fit(im, (96,96))
```

```python
plt.imshow(thumb)
```

```python
plt.imshow(thumb-np.mean(thumb))
```

```python
def preprocessImg(img_file_name,side_length=96):
    im = imread(img_file_name)
    im -= np.mean(im)
    im = Image.fromarray(im)
    thumb = ImageOps.fit(im, (side_length,side_length))
    return np.array(thumb)
```

```python
plt.imshow(preprocessImg('../data/0100448_RP-P-OB-80.995B.jpg'))
```
